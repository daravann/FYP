```python
#composite section B3,B4
I_beam = mono_i_section(
    d=1752.6, b_t=368.3, b_b=368.3, t_ft=15.875, t_fb=19.05, t_w=15.875, r=0, n_r=0
)
panel = sections.rectangular_section(d=190, b=2800, material=concrete)
panel = panel.align_center(I_beam).align_to(I_beam, on="top")
# Create intermediate nodes in panel to match nodes in ub
panel = (panel - I_beam) | panel

section_geometry = CompoundGeometry([I_beam, panel])
section_geometry.create_mesh(mesh_sizes=[30, 30])
comp_section_B3 = Section(section_geometry, time_info=True)
comp_section_B3.display_mesh_info()  # display the mesh information
comp_section_B3.plot_mesh(materials=True, alpha=0.6)
comp_section_B3.calculate_geometric_properties()
# comp_section_B3.display_results(fmt='.3f')
```

```python
a_B3= comp_section_B3.get_area()  
j_B3= comp_section_B3.get_j() 
# Iz_B3 = comp_section_B3.get_ip() 
(Iz_B3,Iy_B3,Ixy_B4)= comp_section_B3.get_ic() 
(Az_B3, Ay_B3)= comp_section_B3.get_As()  
# Ay= comp_section_area *10**-6 
(zxx_plus_B3, zxx_minus_B3, zyy_plus, zyy_minus) = comp_section_B3.get_z()
E_effective = comp_section_B3.get_e_eff()
```
